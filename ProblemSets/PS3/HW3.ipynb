#### Ex5.

#### Ex6.
Finite horizon Bellman equation:
$$V_{T-2}(W_{T-2}) = \max_{\substack{W_{T} \in [0, W_{T-1}] \\ W_{T-1} \in [0, W_{T-2}]}} ln(W_{T-2} - W_{T-1}) + \beta ln(W_{T-1} - W_{T}) + \beta^2 ln(W_{T})$$
Derive the equation by $W_{T-1}$ we get:
$$-\frac{1}{W_{T-2}-W_{T-1}}+\beta\frac{1}{W_{T-1}-W_{T}}=0$$
Then Derive below by $W_T$: 
$$V_{T-1}(W_{T-1})= \max _{W_{T}} \ln(W_{T-1}-W_{T})+\beta V_{T}(W_{T})$$ 
We get:$$-\frac{1}{W_{T-1}-W_{T}}+\beta\frac{1}{W_{T}}=0$$
Solve both equation we get:
$$W_{T-1}=\frac{\beta+1}{\beta}W_T$$
$$W_{T-1}=\psi_{T-2}(W_{T-2})=\frac{\beta+\beta^{2}}{1+\beta+\beta^{2}} W_{T-2}$$
$$V_{T-2}(W_{T-2}) = ln(\frac{W_{T-2}}{1 +\beta +\beta^2}) + \beta ln(\frac{\beta W_{T-2}}{1+\beta+\beta^2}) + \beta^2ln(\frac{\beta^2 W_{T-2}}{1+\beta+\beta^2})$$

#### Ex9.

```python
W_max=1
W_min=0.01
N=100
W=np.linspace(W_min,W_max, N)
W
```

#### Ex10.

```python
N=100
b=0.9
c=np.tile(W.reshape(N,1), (1,N))-np.tile(W, (N,1))
c_pos=c<=0
c[c_pos]=1e-10
u=np.log(c)
v0_m=np.zeros((N, N))
v1=(u+b*v0_m).max(axis=1)
max_id=np.argmax(u+b*v0_m,axis=1)
print(W[max_id])
print(v1)
plt.plot(W,v1)
plt.title("Value Function")
```

#### Ex11.

```python
def dist(v1,v2):
    return (sum((v1-v2)**2))
v0=np.zeros(N)
dist1=dist(v0,v1)
print(dist1)
```

```python
d=10
maxiter=300
dT_s=1e-9
itr=0
v_start=v1
while d>dT_s and itr<=maxiter:
    v_m=np.tile(v_start.reshape((1,N)),(N,1))
    v_m[c_pos]=-9e+4
    v=(u+b*v_m).max(axis=1)
    max_id=np.argmax((u+b*v_m),axis=1)
    d=dist(v,v_start)
    v_start=v
    print('Iteration: ',itr+1, ' distance: ', d)
    itr+=1
plt.plot(W,v)
print("We can see that it takes ", itr, 'iterations.')
```

#### Ex16.

```python
sigma=np.sqrt(0.25)
mu=4*sigma
M=7
e_spc=np.linspace(mu-3*sigma,mu+3*sigma,M)
prb=norm.pdf(e_spc,loc=mu,scale=sigma)
print(prb)
```

#### Ex17.

```python
ue=np.array([u*i for i in e_spc])
v_init=np.zeros((N,M))
ve=np.matmul(v_init,prb.reshape(M,1))
ve_m=np.tile(ve.reshape((1,N)),(N,1))
ve_m[c_pos]=-9e+4
ve_m3=np.array([ve_m for i in range(M)])
vt=ue+b*ve_m3
v_new=np.zeros((N,M))
w_new=np.zeros((N,M))
for i in range(N):
    vt_m=vt[:,i,:]
    v_new[i]=vt_m.max(axis=1)
    idx=np.argmax(vt_m,axis=1)
    w_new[i]=W[idx]
plt.plot(W,np.average(w_new,axis=1))
plt.title("Policy Function")
```

```python
plt.plot(W, np.average(v_new, axis = 1))
plt.title("Value Function")
```

#### Ex18.

```python
def diste(v1,v2):
    return sum(sum((v1-v2)**2))
dist_1=diste(v_new,v_init)
dist_1
```
